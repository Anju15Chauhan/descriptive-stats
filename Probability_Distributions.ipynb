**Poisson Distribution**

```python
lambda_poisson = 1 # average rate(λ)
x_poisson = np.arange(0,n+1)
pmf_poisson = poisson.pmf(x_poisson,lambda_poisson)

print(x_poisson)
print(pmf_poisson)
```

```python
plt.bar(x_poisson,pmf_poisson)
plt.show()
```

```python
lambda_values = [1,4,7]

for lamb in lambda_values:
  x_poisson = np.arange(0,n+1)
  pmf_poisson = poisson.pmf(x_poisson,lamb)
  plt.bar(x_poisson,pmf_poisson,label=f"λ = {lamb}")
  plt.legend()
  plt.xlabel("x")
  plt.ylabel("P(X=x)")

  plt.title(f"Poisson Distribution with λ = {lamb}")

  plt.show()
```

## **Sampling Distributions**

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='whitegrid')
```

```python
df = pd.read_csv("/content/train.csv")
df.head()
```

```python
# Descriptive Statistics
df.describe()
```

```python
# Sample Data
np.random.seed(123)

num_samples = 10
sample_size = 50


sample_means = []

for i in range(num_samples):
  sample = np.random.choice(df['TARGET(PRICE_IN_LACS)'],sample_size,replace = True)
  sample_means.append(sample.mean())
  sns.histplot(sample,kde=True)
  plt.axvline(sample.mean(),color='red',linestyle='--')
  plt.show()
sample_means = np.array(sample_means)
print(sample_means)
```

```python
# Visualize the sampling distribution with sample means

plt.figure(figsize=(10,6))
sns.histplot(sample_means,kde=True)
plt.axvline(df['TARGET(PRICE_IN_LACS)'].mean(),color='red',linestyle='--',label='Population Mean')
plt.axvline(sample_means.mean(),color='green',linestyle='--',label='Sample Mean')
plt.legend()
plt.show()
```

```python
print("Population Mean:",df['TARGET(PRICE_IN_LACS)'].mean())
print("Sample Mean:",sample_means.mean())
```

```python
from os import popen
# Hypothesis
my_hypothesis = 125 # Now constructed mean value
population_mean = df['TARGET(PRICE_IN_LACS)'].mean()

# Sample Stats
sample_mean = sample_means.mean()
sample_std = sample_means.std()

# Calculate Z-score
z_score = (sample_mean - population_mean)/(sample_std)

print("Z-score:",z_score)

# Standard Error
standard_error = sample_std/np.sqrt(500)

print("Standard Error:",standard_error)

# Margin of Error
margin_of_error = np.abs(z_score) * standard_error

print("Margin of Error:",margin_of_error)


# Confidence Intervals
confidence_interval = (sample_mean - margin_of_error,sample_mean + margin_of_error)

print("Confidence Interval:",confidence_interval)
```

```python
# p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print("p-value:",p_value)
print("Z-score:",z_score)
```

```python
# Determine if we reject Null Hypothesis
# Alpha @ 5% error
if p_value < 0.05:
  print("Reject Null Hypothesis")
else:
  print("Fail to Reject Null Hypothesis")
```

## **Problem Statement**: **Predicting Patient Blood Pressure for Healthcare Management**
In a healthcare setting, it is crucial to monitor and predict patient blood pressure (BP) to manage and prevent cardiovascular diseases effectively. Blood pressure is a continuous variable that can take any value within a certain range. The goal is to use continuous probability distributions to model and predict blood pressure readings, thereby aiding in early detection and intervention for at-risk patients.
## Context and Importance
Blood pressure monitoring is vital for managing conditions such as hypertension, which can lead to severe health issues like heart attacks and strokes. By accurately modeling blood pressure readings using continuous probability distributions, healthcare providers can:
- Identify patients at risk of abnormal blood pressure levels.
- Predict future blood pressure trends.
- Implement timely interventions to prevent complications.
