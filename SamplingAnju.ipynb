##**RandomSampling**
 • A simple random sample is a method in which every
memberofthepopulation has an equalchance ofbeing
selected.

 • This can be doneusing a randomnumber generatororby
using a table of randomnumbers.


##**StratifiedSampling**
 * The population is divided into subgroupsbased on
some characteristic and then a random sample is
taken from each stratum.

 * This ensures that the
sample is representative of the population as a
whole.


##**ClusterSampling**
 • In cluster sampling, the population is divided into clusters
or groups, and a random sample of clusters is selected.
Then, all individuals in the selected clusters areincluded in
the sample.

 • This method is useful when it is difficult or expensive to
obtain a list of all individuals in the population.


##**SystematicSampling**
 • In systematic sampling, the population is ordered in some
 way (such as alphabetically or by some characteristic), and
 then arandomstartingpoint is chosen.

 • From that point, every nth individual is selected until the
 desired sample size is reached.


##**ConvenienceSampling**
 • Conveniencesampling involves selecting individuals who
are convenient oreasy to access.

 • This method is oftenused in surveys or studies where the
researcherhas limited time or resources.


##**SamplingDistribution**
 • The sampling distribution is the probability distribution of a
statistic based on a randomsample from a population.

 • It is a theoretical concept used in statistics to understand
the behavior of differentstatistical measures sampled from
a population.

 • It also provides information on the variability of the statistic
across different samples.

 • It allows us to make inferencesabout the population based
on the sample statistics.

1. A researcher wants to estimate the average height of high school students in a city. Describe how they would use random sampling to select 100 students for measurement.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the population size
population_size = 1000

# Calculate the random sampling
sample_size = 100
sample = np.random.choice(population_size, sample_size, replace=False)

# Plot the sampling distribution of the sample means
plt.hist(sample, bins=20)
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()

# Print the sample mean
print("Sample Mean:", np.mean(sample))

# Print the sample standard deviation
print("Sample Standard Deviation:", np.std(sample))
```

9. A school wants to find out the average number of hours students spend on homework. Explain how they would use random sampling to select students from different grades.


```python
# Import necessary libraries
import numpy as np
import random

# Create a hypothetical population of students with their homework hours
# For simplicity, let's assume we have 1000 students and their homework hours are randomly distributed
np.random.seed(42)  # For reproducibility
total_students = 1000
homework_hours = np.random.randint(0, 20, total_students)  # Random hours between 0 and 20

# Determine the sample size
sample_size = 100

# Perform random sampling
sample_indices = random.sample(range(total_students), sample_size)
sampled_hours = homework_hours[sample_indices]

# Calculate the average homework hours for the sampled students
average_homework_hours = np.mean(sampled_hours)

plt.hist(sampled_hours, bins=20)
plt.xlabel('Homework Hours')
plt.ylabel('Frequency')
plt.show()

print(f"Average homework hours based on the sample: {average_homework_hours:.2f} hours per week")
```

10. A public health official wants to estimate the smoking rate among adults in a state. Describe how they would use stratified sampling to include different age groups, genders, and ethnicities.



```python
# Define the population
population = np.random.normal(loc=50, scale=5, size=1000)

# Calculate startified sampling
sample_size = int(len(population) / 10)

startified_samples = [population[i:i+sample_size] for i in range(0, len(population), sample_size)]

# Print the startified samples
for i, sample in enumerate(startified_samples):
    print(f"Startified Sample {i+1}: {sample}")

    # Plot the sampling distribution of the sample means
    plt.hist(sample, bins=20)
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')
    plt.show()

    print("Sample Mean:", np.mean(sample))
    print("Population Mean:", np.mean(population))
    print()
```

19. A transportation authority wants to study commuter patterns. Outline how they would use systematic sampling to survey every 10th passenger on a train.



```python
import numpy as np

# Define the population size
total_passengers = 300

# Generate random data for demonstration (e.g., passenger IDs)
passenger_ids = np.arange(1, total_passengers + 1)

# Determine the sampling interval
interval = 10

# Randomly choose a starting point between 1 and the interval
np.random.seed(42)  # For reproducibility
starting_point = np.random.randint(1, interval + 1)

# Select every 10th passenger starting from the random starting point
selected_passengers = passenger_ids[starting_point - 1::interval]

print("Selected passenger IDs for the survey:")
print(selected_passengers)

plt.hist(selected_passengers, bins=20)
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()
```
