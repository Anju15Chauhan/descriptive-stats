#**Inferential Statistics - Confidence Interval for Mean Salary**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# confidence interval for mean salary
path = '/content/cleaned_employees (1).csv'
df = pd.read_csv(path)
df.head()
```

```python
# mean of the salary
mean_salary = df['Salary'].mean()
print('Mean salary:', mean_salary)

# standard deviation
std_salary = df['Salary'].std()
print('Standard deviation of salary:', std_salary)

# sample size
n = len(df)
print('Sample size:', n)

# confidence level
alpha = 0.05
print('Confidence level:', alpha)

# compute the 95% confidence interval for the mean salary of the employees. Assume that the population standard deviation is unknown. Print the result

import scipy.stats as stats
import numpy as np

# compute the confidence interval
mean_salary = df['Salary'].mean()
std_salary = df['Salary'].std()
n = len(df)
alpha = 0.05
t_critical = stats.t.ppf(1-alpha/2, n-1)
margin_of_error = t_critical * (std_salary / np.sqrt(n))
confidence_interval = (mean_salary - margin_of_error, mean_salary + margin_of_error)
print('95% confidence interval for mean salary:', confidence_interval)
```

#**Hypothesis Testing - Salary Differences Between Departments**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# salary differences between departments
path = '/content/cleaned_employees (1).csv'
df = pd.read_csv(path)
df.head()
```

```python
# average salaries between two departments: Sales and Engineering.
```

```python
# null hypothesis
df_sales = df[df['Department'] == 'Sales']
df_engineering = df[df['Department'] == 'Engineering']
df_sales.head()
df_engineering.head()

```

```python
# Alternative hypothesis
import scipy.stats as stats
t_statistic, p_value = stats.ttest_ind(df_sales['Salary'], df_engineering['Salary'])
print('t-statistic:', t_statistic)
print('p-value:', p_value)

```

```python
# salary of the sales department
mean_sales = df_sales['Salary'].mean()
print('Mean salary of sales department:', mean_sales)

# salary of the engineering department
mean_engineering = df_engineering['Salary'].mean()
print('Mean salary of engineering department:', mean_engineering)
```

```python
# t-test
import scipy.stats as stats
t_statistic, p_value = stats.ttest_ind(df_sales['Salary'], df_engineering['Salary'])
print('t-statistic:', t_statistic)
print('p-value:', p_value)
```

```python
# null hypothesis
h0 = 'There is no significant difference in salaries between the sales and engineering departments.'

# alternative hypothesis
h1 = 'There is a significant difference in salaries between the sales and engineering departments.'

# significance level
alpha = 0.05

# test statistic
t_statistic = stats.ttest_ind(df_sales['Salary'], df_engineering['Salary'])[0]

# critical value
t_critical = stats.t.ppf(1-alpha/2, df_sales['Salary'].count()+df_engineering['Salary'].count()-2)

# conclusion
if t_statistic > t_critical:
  print('Reject null hypothesis')
else:
  print('Fail to reject null hypothesis')
  print(h0)
  print(h1)
  print('There is no significant difference in salaries between the sales and engineering departments.')



```
